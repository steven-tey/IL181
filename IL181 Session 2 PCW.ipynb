```python
data = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")
data.head()        
```

```python
len(data)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

clf = SVC()
X = np.asarray(data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness']])
y = np.asarray(data[["Outcome"]]).ravel()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

clf.fit(X_train, y_train)

predicted_ys = clf.predict(X_test)
accuracy = accuracy_score(predicted_ys, y_test)

print("Accuracy of model trained on the factors 'Pregnancies', 'Glucose', 'BloodPressure', and 'SkinThickness' was ", "{:.2f}".format(accuracy*100), "%")
```

```python
clf = SVC()
X = np.asarray(data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'BMI', 'DiabetesPedigreeFunction', 'Age']])
y = np.asarray(data[["Outcome"]]).ravel()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

clf.fit(X_train, y_train)

predicted_ys = clf.predict(X_test)
accuracy = accuracy_score(predicted_ys, y_test)

print("Accuracy of model trained on all the factors was ", "{:.2f}".format(accuracy*100), "%")
```
